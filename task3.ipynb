EXPLORATORY DATA ANALYSIS

```python
#For multivariate analysis, we can the pairplot()method of seaborn module. 
import seaborn as sns
sns.pairplot(df)
```

```python
#lets create a heat map to understand the relationship in a more better way as few relation are not clearly visible in the abole pairplot .
df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot =True)
plt.show()
```

Looking at the heatmap, we can see that there is a slight positive relation between profit and sales.There is a negative relation between discount & profit as well. 

```python
#get the count of unique values from the region column
df['Region'].value_counts()
```

```python
#visualisation of unique value from the region column through seaborn library
sns.set_style('whitegrid')
sns.countplot(df['Region'], order =df['Region'].value_counts().index)
```

```python
#to get the unique value from the state column
df['State'].value_counts()
```

```python
#visualisation of unique value from the state column
plt.figure(figsize=(18,8))
plt.xticks(rotation=90)
sns.set_style('whitegrid')
sns.countplot(df['State'],order=df['State'].value_counts().index)
```

```python
#to get the count of unique values from the city column
city=df['City'].value_counts()
city.head(10)
```

```python
sns.set_style('whitegrid')
sns.countplot(x=df['Category'], order =df['Category'].value_counts().index)
```

```python
plt.figure(figsize=(14,5))
plt.xticks(rotation=90)
sns.set_style('whitegrid')
sns.countplot(df['Sub-Category'],order=df['Sub-Category'].value_counts().index)
```

```python
sns.set_style('whitegrid')
sns.countplot(df['Segment'])
```

```python
#Visualisation of Sub-Catgory Vs Sales using barplot
plt.figure(figsize=(14,5))
plt.xticks(rotation=90)
sns.set_style('whitegrid')
sns.barplot(x=df['Sub-Category'], y=df['Sales'])
```

```python
#Visualisation of State Vs Pofit using barplot 
plt.figure(figsize=(16,5))
plt.xticks(rotation=90)
sns.set_style('whitegrid')
sns.barplot(x=df['State'], y=df['Profit'])
```

```python
#Visualisation of Top 10 Cities and their Sales
top10cities=df.groupby('City')['Sales'].sum().sort_values(ascending=False).iloc[0:10]
plt.title('Sales of Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Total sum of Sales')
top10cities.plot(kind='bar')
```

```python
#Visualisation of Sales of  Top 10 state w.r.t Quantity
top10states=df.groupby('State')['Quantity'].sum().sort_values(ascending=False).iloc[0:10]
plt.title('Sales of Top 10 State')
plt.xlabel('State')
plt.ylabel('Total sum of Sales')
top10states.plot(kind='bar')
```

```python
#count of discount slabs grouped by city
discount= df.groupby('City')['Discount'].value_counts().sort_values(ascending= False).iloc[0:10]
discount
```

```python
#StateWise sales and profit overview
overview= df.groupby('State')['Profit','Sales'].sum()
overview.plot(kind='bar',figsize=(20,7))
```

So from the graph it is clear that California have maximum sales as well as profit and texas have maximum negative profit.
